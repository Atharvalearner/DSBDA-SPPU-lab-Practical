```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
at = pd.read_csv(r"C:\Users\Shree\Desktop\atharva\DSBDA\datasets\heart.csv")
at
```

```python
plt.scatter(at["age"],at["chol"],color="lightblue")
plt.xlabel('age')
plt.ylabel('chol')
plt.title('age Vs chol')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(at['age'],color='orange',edgecolor='black')
plt.xlabel('age')
plt.ylabel('chol')
plt.title('Distribution of age over chol')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(at[:25]['age'], label='age', linestyle='-', color='green')
plt.xlabel('age')
plt.ylabel('thall')
plt.title('age vs thall')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot([at['age'],at['chol'],at['thalachh']],labels=['age','chol','thalachh'])
plt.title('age vs chol vs thalachh')
plt.xlabel('age vs chol vs thalachh')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
age_counts = {}
for age in at['age'][:10]:
    if age not in age_counts:
        age_counts[age] = 0
    age_counts[age] += 1

ages = list(age_counts.keys())
counts = list(age_counts.values())

plt.figure(figsize=(8, 8))
plt.pie(counts, labels=ages, autopct='%1.1f%%')
plt.title('Occurrences of ages')
plt.show()
```
