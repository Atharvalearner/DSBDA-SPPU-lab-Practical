# matplotlib.pyplot Visualization

```python
import matplotlib.pyplot as plt
```

```python
def bar_plot_total_likes(df):
    post_types = df['Type'].unique()
    total_likes = df.groupby('Type')['Page total likes'].sum()
    
    plt.bar(post_types, total_likes)
    plt.xlabel('Post Type')
    plt.ylabel('Total Likes')
    plt.title('Total Likes by Post Type')
    plt.xticks(rotation=45)
    plt.show()

bar_plot_total_likes(df)
```

```python
def histogram_post_reach(df):
    plt.hist(df['Lifetime Post Total Reach'], bins=20, color='skyblue', edgecolor='black')
    plt.xlabel('Post Reach')
    plt.ylabel('Frequency')
    plt.title('Distribution of Post Reach')
    plt.show()

histogram_post_reach(df)
```

```python
def line_plot_engagement_by_month(df):
    monthly_engagement = df.groupby('Post Month')['Lifetime Engaged Users'].sum()
    plt.plot(monthly_engagement, marker='o')
    plt.xlabel('Month')
    plt.ylabel('Engagement')
    plt.title('Engagement by Month')
    plt.xticks(range(1, 13))
    plt.grid(True)
    plt.show()

line_plot_engagement_by_month(df)
```

```python
def scatter_plot_reach_vs_engagement(df):
    plt.scatter(df['Lifetime Post Total Reach'], df['Lifetime Engaged Users'], alpha=0.5)
    plt.xlabel('Lifetime Post Reach')
    plt.ylabel('Lifetime Engaged Users')
    plt.title('Relationship between Reach and Engagement')
    plt.show()

scatter_plot_reach_vs_engagement(df)
```

```python
def box_plot_reach_by_type(df):
    plt.boxplot([df[df['Type'] == t]['Lifetime Post Total Reach'] for t in df['Type'].unique()], labels=df['Type'].unique())
    plt.xlabel('Post Type')
    plt.ylabel('Lifetime Post Reach')
    plt.title('Distribution of Post Reach by Type')
    plt.show()

box_plot_reach_by_type(df)
```

```python
def pie_chart_post_types(df):
    post_type_counts = df['Type'].value_counts()
    plt.pie(post_type_counts, labels=post_type_counts.index, autopct='%1.1f%%')
    plt.title('Distribution of Post Types')
    plt.show()

pie_chart_post_types(df)
```

```python
def stacked_bar_plot_engagement_metrics(df):
    metrics = ['comment', 'like', 'share']
    metric_data = df.groupby('Type')[metrics].sum()
    metric_data.plot(kind='bar', stacked=True)
    plt.xlabel('Post Type')
    plt.ylabel('Count')
    plt.title('Engagement Metrics by Post Type')
    plt.xticks(rotation=45)
    plt.legend(title='Engagement Metrics')
    plt.show()

stacked_bar_plot_engagement_metrics(df)
```

```python
def pie_chart_paid_posts(df):
    paid_counts = df['Paid'].value_counts()
    plt.pie(paid_counts, labels=paid_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen'])
    plt.title('Distribution of Paid Posts')
    plt.show()

pie_chart_paid_posts(df)
```

```python
from pandas.plotting import scatter_matrix

def scatter_matrix_engagement_metrics(df):
    engagement_metrics = df[['comment', 'like', 'share', 'Total Interactions']]
    scatter_matrix(engagement_metrics, figsize=(10, 10), alpha=0.5)
    plt.suptitle('Scatter Matrix of Engagement Metrics')
    plt.show()

scatter_matrix_engagement_metrics(df)
```

```python
def bar_plot_total_interactions_by_type(df):
    total_interactions = df.groupby('Type')['Total Interactions'].sum()
    plt.bar(total_interactions.index, total_interactions.values, color='skyblue')
    plt.xlabel('Post Type')
    plt.ylabel('Total Interactions')
    plt.title('Total Interactions by Post Type')
    plt.xticks(rotation=45)
    plt.show()

bar_plot_total_interactions_by_type(df)
```

# seaborn Visualization

```python
import seaborn as sns
```

```python
def pair_plot_numerical_variables(df):
    numerical_vars = df.select_dtypes(include=['int64', 'float64'])
    sns.pairplot(numerical_vars)
    plt.show()

pair_plot_numerical_variables(df)
```

```python
def violin_plot_reach_by_type_seaborn(df):
    sns.violinplot(x='Type', y='Lifetime Post Total Reach', data=df)
    plt.xlabel('Post Type')
    plt.ylabel('Lifetime Post Reach')
    plt.title('Distribution of Post Reach by Type')
    plt.show()

violin_plot_reach_by_type_seaborn(df)
```

```python
def box_plot_reach_by_month_and_type(df):
    sns.boxplot(x='Post Month', y='Lifetime Post Total Reach', hue='Type', data=df)
    plt.xlabel('Post Month')
    plt.ylabel('Lifetime Post Reach')
    plt.title('Post Reach by Month and Type')
    plt.legend(title='Post Type')
    plt.show()

box_plot_reach_by_month_and_type(df)
```

```python
def heatmap_engagement_correlation_seaborn(df):
    engagement_metrics = df[['comment', 'like', 'share', 'Total Interactions']]
    correlation_matrix = engagement_metrics.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation between Engagement Metrics')
    plt.show()

heatmap_engagement_correlation_seaborn(df)
```

```python
def count_plot_post_types(df):
    sns.countplot(x='Type', data=df)
    plt.xlabel('Post Type')
    plt.ylabel('Count')
    plt.title('Distribution of Post Types')
    plt.show()

count_plot_post_types(df)
```

```python
def joint_plot_likes_vs_shares(df):
    sns.jointplot(x='like', y='share', data=df, kind='scatter')
    plt.xlabel('Likes')
    plt.ylabel('Shares')
    plt.title('Likes vs Shares')
    plt.show()

joint_plot_likes_vs_shares(df)
```

```python
def facet_grid_post_reach_by_type(df):
    g = sns.FacetGrid(df, col='Type', col_wrap=3)
    g.map(plt.hist, 'Lifetime Post Total Reach', bins=20, color='skyblue')
    g.set_axis_labels('Post Reach', 'Frequency')
    plt.suptitle('Distribution of Post Reach by Type')
    plt.show()

facet_grid_post_reach_by_type(df)
```

```python
def kde_plot_post_reach(df):
    sns.kdeplot(df['Lifetime Post Total Reach'], shade=True)
    plt.xlabel('Post Reach')
    plt.ylabel('Density')
    plt.title('Distribution of Post Reach with KDE')
    plt.show()

kde_plot_post_reach(df)
```

```python
def bar_plot_mean_reach_by_type(df):
    sns.barplot(x='Type', y='Lifetime Post Total Reach', data=df, ci='sd')
    plt.xlabel('Post Type')
    plt.ylabel('Mean Lifetime Post Reach')
    plt.title('Mean Post Reach by Type with Error Bars')
    plt.show()

bar_plot_mean_reach_by_type(df)
```

```python
def strip_plot_reach_by_month(df):
    sns.stripplot(x='Post Month', y='Lifetime Post Total Reach', data=df, jitter=True)
    plt.xlabel('Post Month')
    plt.ylabel('Lifetime Post Reach')
    plt.title('Distribution of Post Reach by Month')
    plt.show()

strip_plot_reach_by_month(df)
```

```python
import seaborn as sns

def histogram_post_reach_seaborn(df):
    sns.histplot(df['Lifetime Post Total Reach'], kde=True)
    plt.xlabel('Post Reach')
    plt.ylabel('Frequency')
    plt.title('Distribution of Post Reach with Seaborn')
    plt.show()

histogram_post_reach_seaborn(df)
```
